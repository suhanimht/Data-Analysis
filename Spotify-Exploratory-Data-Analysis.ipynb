```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df_tracks= pd.read_csv("C:/Users/Administrator/Documents/Personal/tracks.csv")
df_tracks.head()
```

```python
#Check for null values
pd.isnull(df_tracks).sum()
```

```python
#Check information about the columns and datas
df_tracks.info()
```

```python
df_tracks.describe().transpose()
```

```python
#Find the top 20 least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(20)
sorted_df
```

```python
#Find top 20 most popular songs with popularity > 90
most_popular = df_tracks.query('popularity >90', 
                              inplace = False).sort_values('popularity', 
                                                           ascending = False)
most_popular[:20]
```

```python
df_tracks[['artists']].iloc[18]
```

```python
#convert duration from ms to s
df_tracks['duration'] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace= True, axis=1)
df_tracks.duration.head()
```

```python
average_popularity= df_tracks.groupby('artists')['popularity'].mean()
sorted_artists = average_popularity.sort_values(ascending=False)
top_artist = sorted_artists.index[0]
top_artist
```

```python
#drop unwanted cols
# Drop non-numeric columns
non_numeric_cols = ['id','name','artists','id_artists','key', 'mode', 'explicit']
df_numeric = df_tracks.drop(non_numeric_cols, axis=1)

# Calculate correlation matrix
corr_df = df_numeric.corr(method='pearson')

# Plot the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title("Correlation heatmap of features")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#Creating a sample data (0.5% of original data) from the whole dataset for modelling
sample_df = df_tracks.sample(int(0.005*len(df_tracks)))
print(len(sample_df))
```

```python
#Regression model between loudness and Energy
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y='loudness', x='energy', color='b').set(title= "Loudness vs Energy correlation")
```

```python
#Regression model between Popularity and Acousticness
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y='popularity', x='acousticness', color='r').set(title= "Popularity vs Acousticness correlation")
```

```python
#strip to get dates as years
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates= pd.to_datetime(df_tracks.dates, format='mixed')
years=df_tracks.dates.dt.year
```

```python
total_duration = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x= years, y=total_duration, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation =90)
```

```python
#Line plot to see relation between the total duration of songs in each year
total_duration=df_tracks.duration 
sns.set_style(style="whitegrid")
fig_dims=(10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_duration, ax=ax).set(title="Year vs Duration") 
```

```python
#Importing new dataset to learn about the genres on spotify 
df_genres=pd.read_csv('C:/Users/Administrator/Documents/Personal/SpotifyFeatures.csv')
df_genres.head()
```

```python
#Find duration of songs in different genres
plt.title('Duration of songs in different genres')
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y='genre', x='duration_ms', data=df_genres)
plt.xlabel('Duration in ms')
plt.ylabel('Genre')
```

```python
#Top 5 popular genres
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous= df_genres.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='genre', x='popularity', data = famous).set(title="Top 5")
```
